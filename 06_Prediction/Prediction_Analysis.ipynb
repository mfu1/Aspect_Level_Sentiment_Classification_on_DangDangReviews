```python
import pandas as pd
import numpy as np
```

```python
labels = ['negative', 'positive']
label_mapping = {labels[i]:i for i in range(len(labels))}
label_mapping_rev = {i:labels[i] for i in range(len(labels))}
```

```python
dataset_aspect = pd.read_csv("../05_Transfer/data/Sentiment_Classification/dataset_test_keep.csv", "\t", header=None)
dataset_aspect.columns = ["label", "aspect", "text", "pidx"]
```

```python
pred_without_aspect = pd.read_csv("Predicted_Sentiment_Without_Aspect/test_results.tsv", "\t", header=None)
pred_without_aspect["pred"] = np.argmax(pred_without_aspect.values, axis=1)
```

```python
pred_with_aspect = pd.read_csv("Predicted_Sentiment_With_Aspect/test_results.tsv", "\t", header=None)
pred_with_aspect["pred"] = np.argmax(pred_with_aspect.values, axis=1)
```

```python
dataset_aspect["pred_without_aspect"] = pred_without_aspect["pred"].apply(lambda x:label_mapping_rev[x])
dataset_aspect["pred_with_aspect"] = pred_with_aspect["pred"].apply(lambda x:label_mapping_rev[x])
```
