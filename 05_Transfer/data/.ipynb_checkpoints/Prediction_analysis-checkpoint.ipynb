```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("DangDangReview__test_results.tsv", "\t", header=None)
```

```python
data["max_prob_idx"] = np.argmax(data.values, axis=1)
```

```python
aspect_mapping = {i:j for i, j in enumerate(['功能', '品质', '设计', '使用', '服务', '价格'])}
```

```python
data["aspect"] = data["max_prob_idx"].apply(lambda x:aspect_mapping[x])
```
